# Aula 2.2 - O funcionamento da Árvore de Decisão

    1.0 Introdução a Árvore de Decisão
    A Árvore de Decisão é um algoritmo de Machine Learning utilizado tanto para
    regressão quanto para classificação, na solução de problemas complexos. O funcionamento da Árvore de Decisão é baseado em dividir o conjunto de dados em subconjuntos menores, de forma recursiva, até chegar a um conjunto de dados homogêneo ou até o limite máximo de crescimento ser atingido. Para isso, é necessário escolher a melhor variável para dividir o conjunto de dados em cada etapa, utilizando critérios como Gini, Entropia ou Ganho de Informação.
