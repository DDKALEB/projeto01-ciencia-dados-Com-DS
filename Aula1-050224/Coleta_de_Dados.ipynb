```python
# 2. Selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna distance é maior do que 2000, da tabela
# “flight_activity”

# consulta dos dados no banco de dados
consulta_atividade = """

  SELECT
    flj.loyalty_number,
    flj.year, month,
    flj.flights_booked,
    flj.total_flights,
    flj.distance,
    flj.points_accumulated
  FROM flight_activity flj
    WHERE
      flj.distance > 2000
  """
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```

```python
# 3. Selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna distance é maior do que 2000 e a coluna
# month é igual a 9, da tabela “flight_activity”

# consulta dos dados no banco de dados
consulta_atividade = """

  SELECT
    flj.loyalty_number,
    flj.year, month,
    flj.flights_booked,
    flj.total_flights,
    flj.distance,
    flj.points_accumulated
  FROM flight_activity flj
    WHERE
      flj.distance > 2000 AND flj.month = 9
  """
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```

```python
# 4. Selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna distance é maior do que 2000 ou a coluna
# points_accumulated é menor que 100, da tabela “flight_activity”.

# consulta dos dados no banco de dados
consulta_atividade = """

  SELECT
    flj.loyalty_number,
    flj.year, month,
    flj.flights_booked,
    flj.total_flights,
    flj.distance,
    flj.points_accumulated
  FROM flight_activity flj
    WHERE
      flj.distance > 2000 OR points_accumulated < 100
  """
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```

```python
# 5. Selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna loyalty_card é igual a Star da tabela
# “flight_loyalty_history”.

# consulta dos dados no banco de dados

# Quando Vc não souber quais colunas tem na tabela, para fazer a consulta Vc usa o coringa "*"
consulta_atividade = """

  SELECT
    *
  FROM flight_loyalty_history flh
  WHERE
      flh.loyalty_card = "Star"
  """
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```

```python
# 2. Selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna gender é igual a Female da tabela
# “flight_loyalty_history”.

# consulta dos dados no banco de dados
consulta_atividade = """

SELECT
  flh.loyalty_number,
  flh.country,
  flh.city,
  flh.gender,
  flh.loyalty_card
  FROM
  flight_loyalty_history flh
WHERE
  flh.gender = 'Female'
"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )
# visualiza os dados
df_atividade.head()
```
